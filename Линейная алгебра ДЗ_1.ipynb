##__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

а) (0, a, b) * (c, 0, d) = (c, a, b+d)
Получившийся вектор не имеет 0 ни в первой, ни во второй координате. Соответственно, данная совокупность векторов не является линейным подпространством $$R^3$$

б) $$a = a_{1}*u_{1} + a_{2}u_{2}+ ... +a_{n}*u_{n} = 0  $$
$$b = b_{1}*u_{1} + b_{2}u_{2}+ ... +b_{n}*u_{n} = 0  $$

$$a + b = (a_{1}+b_{1})*u_{1} +(a_{2}+b_{2})*u_{2} +...+ (a_{n}+b_{n})*u_{n}$$

$$k*a = k*a_{1}*u_{1} +k*a_{2}*u_{2} +...+ k*a_{n}*u_{n}$$

Везде получаем векторыиз того же линейного подпространства, значит множество этих вектоов является линейным подпространством
